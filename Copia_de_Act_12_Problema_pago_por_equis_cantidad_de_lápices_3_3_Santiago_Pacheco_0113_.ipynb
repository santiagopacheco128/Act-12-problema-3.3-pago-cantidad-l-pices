## Las variables
1. x --> Cantidad de lapices
2. PAG --> Pago que se realiza por los lapices
___
## Formulas
* $ PAG = X * 0.85 $
* $ PAG = X * 0.90 $
___
## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/santiagopacheco128/Act-12-problema-3.3-pago-cantidad-l-pices/refs/heads/main/DF%20lapices.png" width="300">





```python
# para cuando es verdadero (si)
# calcular el pago usando if else
# datos de entrada x
# mostrar el pago segun la decision
x = float(input("Ingrese la cantidad de lapices: "))
if x >= 1000:
    PAG = x * 0.85
else:
    PAG = x * 0.90
print("El pago a realizar es: ", PAG)
print("Fin del programa")
print("programa realizado por Santiago Pacheco NL = 43")
```

```python
# para cuando es falso (no)
# calcular el pago usando if else
# datos de entrada x
# mostrar el pago segun la decision
x = float(input("Ingrese la cantidad de lapices: "))
if x < 1000:
    PAG = x * 0.90
else:
    PAG = x * 0.85
print("El pago a realizar es: ", PAG)
print("Fin del programa")
print("programa realizado por Santiago Pacheco NL = 43")
```
